*Probabilidades*



La estadística y la teoría de la probabilidad constituyen una rama de las matemáticas para tratar la incertidumbre. La teoría de la probabilidad proporciona una base para la ciencia de la inferencia estadística a partir de datos

1. Muestra: (de tamaño n) obtenida de una población madre que se supone que está representada por una probabilidad

2. Estadística descriptiva: descripción de la muestra

3. Estadística inferencial: tomar una decisión o una inferencia a partir de una muestra de nuestro problema

*Valores de probabilidad*

Un conjunto de valores de probabilidad para un experimento con espacio muestral $S={O1,02,⋯,On}$ consta de algunas probabilidades que satisfacen:
$0≤pi≤1, i=1,2,⋯,n$ y $p1+p2+⋯+pn=1$

La probabilidad de que ocurra el resultado Oi se dice que es pi y se escribe:
$P(Oi)=pi$
En los casos en los que los n resultados son igualmente probables, entonces cada probabilidad tendrá un valor de 1/n.

*Eventos y complementos*

*Eventos: subconjunto del espacio muestral

La probabilidad de un evento A, P(A), se obtiene a partir de las probabilidades de los resultados contenidos en el evento A

Se dice que un evento ocurre si ocurre uno de los resultados contenidos en el evento

Complemento de eventos: el evento A′ es el evento que consiste en todo lo que hay en el espacio muestral S que no está contenido en A
$P(A)+P(A')=1$

EventO simple: evento que consiste en un resultado individual

*Combinación de eventos*



1.   Intersección.
A⋂B consiste en los resultados contenidos en ambos eventos A y B.

Probabilidad de intersección. P(A⋂B) es la probabilidad de que los eventos ocurran simultáneamente.

Propiedades:
* P(A⋂B)+P(A⋂B')=P(A)
* Eventos mutuamente excluyentes: si A⋂B = ∅
* A⋂(B⋂C)= (A⋂B)⋂C

2. Unión. A⋃B consiste en los resultados que están contenidos en al menos uno de los eventos A y B.

La probabilidad de este evento, P(A∪B) es la probabilidad de que al menos uno de estos eventos A y B ocurra.

Propiedades:
* P(A⋃B)= P(A⋂B') + P(A'⋂B') + P(A⋂B)
*Si son eventos mutuamente exclusivos: P(A⋃B)= P(A)+P(B)

**Combinación de tres o más eventos**

*Probabilidad condicional*

Probabilidad condicional: de un evento A condicional a un evento B es:
$P(A|B)= \frac{P(A⋂B)}{P(B)}$ para P(B)>0

*Eventos independientes*

Dos eventos son independientes si:
* P(A|B)= P(A)
*P(B|A)= P(B)
*P(A⋂B)= P(A)P(B)

Interpretación: los eventos son independientes si el conocimiento sobre un evento no afecta la probabilidad del otro evento.

*Ley total de la Probabilidad*

Dada {A1, A2,..., An}, una partición del espacio muestral S, la probabilidad de un evento B, P(B)se puede expresar como:
$P(B)= ∑_{i=1}^{n}P(A_{i})P(B|A_{i})$

**Teorema de Bayes**

Dada {A1, A2, ..., An}, una partición un espacio muestral, entonces las probabilidades posteriores del evento A_{i} condicional a un evento B se pueden obtener a partir de las probabilidades P(A_{i})vy P(A_{i}|B) usando la fórmula:
$P(A_{i}|B)= \frac{P(A_{i})P(B|A_{i})}{∑_{j=1}^{n}P(A_{j})P(B|A_{j})}$

La representación más común de una distribución es un histograma, que es un gráfico que muestra la frecuencia o probabilidad de cada valor.

```python
#Histograma
t=([1,2,2,3,1,2,3,2,1,3,3,3,3])
#Frecuencia
hist={}
for x in t:
 hist[x]=hist.get(x,0)+1
 #Normalización
 n= float(len(t))
 pmf={}
 for x, freq in hist.items():
  pmf[x]=freq/n

print("Histograma:", hist)
print("PMF:", pmf)
```

En esta parte normalizaremos los datos de las siguientes universidades, donde notamos que Cornell tiene la peor proporción de alumno por profesor. La normalización es el proceso de convertir estas frecuencias en probabilidades, diviendo cada frecuencia por n, donde n es el tamaño de nuestra lista original.

```python
# Datos de universidades
universities = {
    "Dartmouth": {"students": 5000, "professors": 300},
    "Columbia": {"students": 11000, "professors": 500},
    "Brown": {"students": 8000, "professors": 400},
    "Cornell": {"students": 16000, "professors": 650}
}

# Histograma de estudiantes y profesores
students_hist = {uni: data["students"] for uni, data in universities.items()}
prof_hist = {uni: data["professors"] for uni, data in universities.items()}

# Total de estudiantes y profesores
total_students = sum(students_hist.values())
total_prof = sum(prof_hist.values())

# Normalización para calcular las proporciones (PMF)
student_pmf = {uni: students / total_students for uni, students in students_hist.items()}
prof_pmf = {uni: professors / total_prof for uni, professors in prof_hist.items()}

# Resultados
print('Proporciones de estudiantes (PMF):', student_pmf)
print('Proporciones de profesores (PMF):', prof_pmf)
```

*Distribución Binomial y Distribuición de Bernoulli*
Una distribución binomial tiene un número contable de resultados y, por lo tanto, es discreta.

Una distribución de Bernoulli es un caso especial de distribución binomial.

Todos los valores en una distribución de Bernoulli son 0 o 1.

Por ejemplo, si tomamos una moneda no justa que cae de cara el 60 % de las veces, podemos describir la distribución de Bernoulli de la siguiente manera:

p (cambio de caras) = ​​0.6

1 - p (cambio de cruces) = 0.4

caras = 1

cruces = 0

Formalmente, podemos describir una distribución de Bernoulli con la siguiente FMP:


```python
#Cambio de caras:
p= 0.6
#Muestras de Bernoulli:
bern_dist = bernoulli.rvs(p, size=1000)
bern_df= pd.DataFrame({'value' : bern_dist})
bern_values = bern_df['value'].value_counts()

#Gráfica de ditribución
plt.figure(figsize=(10,5))
bern_values.plot(kind='bar', rot=0)
plt.annotate(xy=(0.85,300), text='Muestras de Tails/nn= {}'.format(bern_values[0]), fontsize= 'medium', color= 'white')
plt.annotate(xy=(-0.2,300), text="Muestras de Heads/nn={}".format(bern_values[1]), fontsize='large', color='white')
plt.title('Distribución de Bernoulli con p=0.6, n=1000')
plt.show()
```

**MSE Y RMSE**

El error cuadrático medio (MSE, por sus siglas en inglés) es una fórmula para medir cuánto se desvían los estimadores de la distribución real. Esto puede ser muy útil, por ejemplo, para evaluar modelos de regresión.

El error cuadrático medio (RMSE, por sus siglas en inglés) es simplemente la raíz del MSE.

```python
from math import sqrt
Y= 100 #Valor actual
YH= 94 #Valor predecido

#MSE
def MSE(Y, YH):
  return np.square(YH-Y).mean()

#RMSE
def RMSE(Y, YH):
  return sqrt(np.square(YH-Y).mean())

print('MSE:', MSE(Y, YH))
print('RMSE:', RMSE(Y,YH))
```

*Regresión Lineal*

La regresión lineal se puede realizar mediante el método de mínimos cuadrados ordinarios (MCO) o la estimación de máxima verosimilitud (MLE).

La mayoría de las bibliotecas de Python utilizan MCO para ajustar modelos lineales.

```python
# Datos
x = np.random.uniform(low=20, high=260, size=100)
y = 50000 + 2000*x - 4.5 * x**2 + np.random.normal(size=100, loc=0, scale=10000)

#Grafica de datos con regresión lineal
plt.figure(figsize=(16,5))
plt.title('Regresión Lineal con datos cuadráticos', fontsize='xx-large')
sns.regplot(x=x, y=y)
plt.show()
```

```python
#Regresión lineal con scratch
import random
#Crear datos
xs = np.array(range(1,20))
ys = [0,8,10,8,15,20,26,29,38,35,40,60,50,61,70,75,80,88,96]
#Agregar datos en diccionario
data= dict()
for i in range(len(xs)):
  data.update({xs[i-1]:ys[i-1]})

#Pendiente
m=0
#y
b=0
#Rate de aprendizaje
lr=0.0001
#Numero de epochs
epochs=10000
#fórmula lineal
def line(x):
  return m*x+b
#Algoritmo de regresión lineal
for i in range(epochs):
  #Dato aleatorio para calcular la distancia horizontal y vertical
 rand_point = random.choice(list(data.items()))
 vert_dist = abs((m * rand_point[0] + b) - rand_point[1])
 hor_dist = rand_point[0]

 if (m * rand_point[0] + b) - rand_point[1] < 0:
 #Ajustar linea superior
  m += lr * vert_dist * hor_dist
  b += lr * vert_dist
 #Ajustar linea inferior
 else:
  m -= lr * vert_dist * hor_dist
  b -= lr * vert_dist
#Graficamos
plt.figure(figsize=(15,6))
plt.scatter(data.keys(), data.values())
plt.plot(xs, line(xs))
plt.title('Resultado de regresión lineal')
print('Pendient: {}\nIntercept: {}'.format(m, b))
```

Esta es la probabilidad de encontrar resultados iguales o más extremos cuando la hipótesis nula (H0) es verdadera. En otras palabras, un valor p bajo significa que tenemos evidencia convincente para rechazar la hipótesis nula.

Si el valor p es inferior al 5 % (p < 0,05), a menudo rechazamos H0 y aceptamos que Ha es verdadera. Decimos que p < 0,05 es estadísticamente significativo, porque hay menos del 5 % de posibilidades de que nos equivoquemos al rechazar la hipótesis nula.

*Método Tukey*

Tukey sugirió que una observación es un valor atípico cuando una observación es 1,5 veces el rango intercuartil por debajo del primer cuartil o 1,5 veces el rango intercuartil por encima del tercer cuartil. Esto puede parecer complicado, pero es bastante intuitivo si lo vemos visualmente.

Para distribuciones normales, los criterios de Tukey para observaciones atípicas son poco probables si no hay valores atípicos presentes, pero el uso de los criterios de Tukey para otras distribuciones debe tomarse con cautela.

```python
#Detectar outliers en la columna 'Income' de la base de datos.
#Definimos la función
def tkey_outliers(data):
 th=3

 promedio= np.mean(data)
 desviacion_estandar= np.std(data)
#Recolectar outliers
 outliers = []
 for i in data:
  z_score = (i - promedio) / desviacion_estandar
  th=3
  if abs(z_score) > th:
   outliers.append(i)
 return outliers

 #Salida de outliers
 income_outliers = turkey_outliers(df['Income'])
 #Gráfica
plt.figure(figsize=(15,6))
df['Income'].plot(kind='hist', bins=100, label='Income distribución')
plt.hist(income_outliers, bins=20, label='Outliers')
plt.title("Outlier detección con el método de Tukey", fontsize='xx-large')
plt.xlabel('Income')
plt.legend(fontsize='xx-large')
```

*Función de enlace*

Una función de enlace se utiliza en los modelos lineales generalizados (GLM) para aplicar modelos lineales a una variable de respuesta continua dada una serie de predictores continuos y/o categóricos. Una función de enlace que se utiliza a menudo se denomina función logit inversa o función sigmoidea logística.

La función de enlace proporciona una relación entre el predictor lineal y la media de una distribución.

```python
#Función logit inversa (funcion de enlace)
def inv_logit(x):
  return 1/(1+np.exp(-x))
t1=np.arange(-10,10,0.1)
plt.figure(figsize=(10,5))
plt.plot(t1, inv_logit(t1),
         t1, inv_logit(t1-2),
         t1, inv_logit(t1*2))
plt.title('Función logit inversa', fontsize='xx-large')
plt.legend(('Normal', 'Cambio de intercept', 'Cambio de pendiente'), fontsize='xx-large')
```

*Regresión logística*

Con la regresión logística, utilizamos una función de enlace como la función logit inversa mencionada anteriormente para modelar una variable dependiente binaria. Mientras que un modelo de regresión lineal predice el valor esperado de y dado x directamente, un GLM utiliza una función de enlace.

Podemos implementar fácilmente la regresión logística con la función de regresión logística de sklearn.

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
X,y = load_iris(return_X_y=True)

#Clasificador
clf= LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(X,y)

print('Presición del modelo de regresión logística sobre la base de datos de Iris flower: {} '.format(clf.score(X,y)))
```

*Prueba de hipótesis: comparación de dos grupos*

scipy.stats.ttest_1samp()Comprueba si es probable que la media de la población de los datos sea igual a un valor dado (técnicamente, si las observaciones se extraen de una distribución gaussiana de una media de población dada). Devuelve la estadística T y el valor p.

```python
from scipy import stats
print('Test 1')
print(stats.ttest_1samp(data['VIQ'], 0))

#Prueba t de 2 muestras: prueba de diferencias entre poblaciones
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
print('Test 2')
print(stats.ttest_ind(female_viq, male_viq))

#Pruebas pareadas: mediciones repetidas en los mismos individuos
#PIQ, VIQ y FSIQ proporcionan tres medidas del CI. Probemos
#si FISQ y PIQ son significativamente diferentes
print('Test 3')
stats.ttest_rel(data['FSIQ'], data['PIQ'])
#Eliminamos los vínculos entre las repeticiones con una prueba de medidas repetidas
print('Test 3 con prueba de medidas repetidas')
stats.ttest_rel(data['FSIQ'], data['VIQ'])

print('Prueba de una muestra sobre la diferencia')
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)

#Las pruebas t suponen errores gaussianos.
# Podemos utilizar una prueba de rangos con signo de Wilcoxon

print('Prueba de Wilcoxon')
stats.wilcoxon(data['FSIQ'], data['PIQ'])
```

**Modelos Lineales**

*Regresión Lineal simple*

Statsmodels utiliza una terminología estadística: la variable y se denomina "endógena", mientras que la variable x se denomina "exógena".

Para simplificar, y (endógeno) es el valor que estás intentando predecir, mientras que x (exógeno) representa las características que estás utilizando para hacer la predicción.



```python
from numpy import random
x= np.linspace(-5, 5, 20)
np.random.seed(1)

y=-5+3*x+4*random.normal(size=x.shape)
#Creamos un dataframe con las variables relevantes
data= pd.DataFrame({'x': x, 'y': y})
#sns.regplot(x='x', y='y', data=data)

#Aplicamos un modelo MCO y lo ajustamos:
from statsmodels.formula.api import ols
model= ols("y~x", data).fit()
print(model.summary())  #Regresión lineal simple
```

*Variables Categóricas: comparación de grupos o categorías múltiples*

```python
#Podemos escribir una comparación entre el CI de hombres y mujeres utilizando un modelo lineal.
#Se puede forzar que una columna de números enteros se trate como categórica utilizando:
model= ols('VIQ~C(Gender)', data).fit()
print(model.summary())
```

```python
#Enlace a pruebas t diferentes entre FSIQ y PIQ
#Para comparar diferentes tipos de CI, necesitamos crear una tabla de
#“formato largo”, que enumere los CI:
data_fisq = pd.DataFrame({'iq': data['FSIQ'], 'type': 'fsiq'})
data_piq = pd.DataFrame({'iq': data['PIQ'], 'type': 'piq'})
data_long = pd.concat((data_fisq, data_piq))
print(data_long)
model = ols("iq ~ type", data_long).fit()
print(model.summary())
#Notemos que recuperamos los valores anteriores de la prueba t y valor p:
stats.ttest_ind(data['FSIQ'], data['PIQ'])
```

```python
#Datos de prueba
x = np.array([59, 47, 44, 32, 45, 61, 71,
              32, 28, 36, 36, 51, 31, 44,
              73, 47, 40, 43, 68, 40])
y = np.array([23, 45, 29, 55, 48, 42, 32,
              33, 34, 52, 40, 35, 45, 47,
              26, 42, 35, 41, 29, 38])
#Creamos un dataframe:
data = pd.DataFrame({'x': x, 'y': y})
#Data fitting
fit= ols('y~x', data).fit()
print(fit.summary())
#Graficas
abline_plot(model_results=fit)
plt.plot(x, y, 'bo')
plt.show()
#Coeficientes
#display(Math(r'\hat{\beta_0} = {}'.format(fit.params[0])))
#display(Math(r'\hat{\beta_1} = {}'.format(fit.params[1])))
display(Math(r'\hat \beta_0: {:.4f} \\ \hat \beta_1: {:.4f}'.format(
    fit.params[0], fit.params[1])))
```

```python
'''Ejemplo teniendo las sumas x,y,x^2,y^2'''
#Parámetros:
n=20
sum_x=8.552
sum_y=398.2
sum_x2=5.196
sum_y2= 9356
sum_xy=216.6

#Cálculos
x_bar=sum_x/n
y_bar=sum_y/n
Sxx=(sum_xy-n * x_bar * y_bar)
Sxy= (sum_x2-n * x_bar **2)
beta_1= Sxx/Sxy
beta_0=y_bar-beta_1*x_bar
SSE= sum_y2-beta_0*sum_y - beta_1*sum_xy
sigma2= SSE/(n-2)

#Resultados
display(Math(r'\beta_0 = {:.4f}\\ \beta_1 = {:.4f}'.format(beta_0, beta_1)));
display(Math(r'SSE = {:.4f}\\ \sigma^2 = {:.4f}'.format(SSE, sigma2)));
#Predicción: si x=0.5, cuánto vale y?
x=0.5
y=beta_0+beta_1*x
print('y={:.4f}'.format(y))
```

```python
#Parámetros
n = 24
alpha = 0.01
beta_hat = 0.8051
se_beta = 0.0643 #error estándar
beta_10 = 0

#Cálculos
df= n-2
critical = t.ppf(1-alpha/2, df)
left = beta_hat - critical * se_beta
right = beta_hat + critical * se_beta

#T-estadisticas
t_stat= (beta_hat-0)/ se_beta
p_value= t.sf(t_stat, df=df)

print('Punto crítico: {:.4f}'.format(critical))
print('Intervalo de confianza: [{:.4f}, {:.4f}]'.format(left, right))
print('p-value: {:.4f}'.format(p_value))
```

Análisis de Tabla de Varianzas

```python
#Dataset
x = np.array([52, 68, 64, 88, 80, 75,
              59, 63, 85, 74, 71, 66])

y = np.array([64, 53, 58, 59, 49, 54,
              38, 48, 68, 63, 58, 47])
#Dataframe
data = pd.DataFrame({'x': x, 'y': y})
#fit model
fit= ols('y~x', data).fit()

#Tabla ANOVA
anova_table= anova_lm(fit)
print('\n Tabla Anova'); print(anova_table)

SSE= anova_table['sum_sq'][0]
SSR= anova_table['sum_sq'][1]

#Calcular R^2
R2 = SSE/(SSR+SSE)
display(Math(r'R^2 = {:.4f}'.format(R2)))
```

*Análisis de correlación*

```python
#dataset
x = np.array([60, 61, 63, 63, 65, 65, 65, 66,
              68, 68, 69, 70, 70, 72, 72, 72,
              74, 75, 77, 77, 80, 80, 81, 81])

y = np.array([59.47, 63.43, 61.59, 62.44, 65.84,
              64.21, 66.87, 64.82, 68.93, 69.71,
              69.80, 67.95, 71.14, 68.71, 71.64,
              73.06, 70.73, 74.57, 70.86, 76.09,
              73.89, 78.39, 76.87, 81.53])
#covarianza
def cov(X, Y):
  x_bar= X.mean()
  y_bar= Y.mean()
  return sum((X-x_bar)*(Y-y_bar))
#Desviación estándar
def sigma(X):
  x_bar= X.mean()
  return math.sqrt(sum((X-x_bar)**2))

r= cov(x, y)/(sigma(x)*sigma(y))
print('Coeficiente de correlación (cálculo manual): {:.4f}'.format(r))
print('Correlación de coeficiente (Vía numpy): {:.4f}'.format(np.corrcoef(x,y)[0,1]))

n=len(x)
t_stat_1= (r*math.sqrt(n-2))/(math.sqrt(1-r**2))
print('t-estadistica con método: {:.4f}'.format(t_stat_1))
```

c) ¿Qué tan grande debe elegir n para que la diferencia entre estas dos funciones sean menores que 0.1?

```python
#Definiremos una función para calcular la diferencia máxima entre dos arrays
def max_difference(y1, y2):
    return np.max(np.abs(y1 - y2))

# Inicializar n
n = 10
epsilon = 1 / 5
diferencia_maxima = 0.1  # Condición de la diferencia máxima

# Bucle para aumentar n hasta que la diferencia sea menor que 0.1
while True: #Usamos un ciclo while para incrementar el valor de n
    n2 = n + 10 #así la diferencia máxima entre las dos funciones es menor a 0.1

    # Crear nodos
    x1 = np.linspace(0, 1, n + 1)
    x2 = np.linspace(0, 1, n2 + 1)

    # Calcular las funciones
    y1 = f(x1, epsilon)
    y2 = f(x2, epsilon)

    # Interpolación para comparar en los mismos puntos
    y2_interp = np.interp(x1, x2, y2)

    # Calcular la diferencia máxima
    diff = max_difference(y1, y2_interp)

    # Verificar si la diferencia es menor que 0.1
    if diff < diferencia_maxima:
        break

    # Incrementar n
    n += 1

# Imprimir el valor de n adecuado
print(f'El valor adecuado para n debe ser de: {n}')
```

a)Desarrollar un programa que cree la gráfica de la derivada de f= f(x) y la derivada exacta dada por f'=f'(x)
b) Probar el programa usando n=10 y ε=1/5

```python
# Anteriormente habíamos definido f(x)

# Definimos la derivada exacta de f(x)
def f_exacta(x, epsilon):
    return -np.cos(1 / (x + epsilon)) / (x + epsilon)**2

# Derivada aproximada de f(x) usando diferencias finitas
def f_approx(x, epsilon, h=1e-5):
    return (f(x + h, epsilon) - f(x, epsilon)) / h

# Parámetros iniciales
epsilon = 1 / 5
n = 10  # número de nodos
x = np.linspace(0, 1, n + 1)[1:]  # Dividimos el intervalo [0,1], pero excluimos 0 para evitar singularidad en x = 0
h = x[1] - x[0]  # Espaciado entre nodos

# Calcular la derivada exacta y la derivada aproximada
f_exacta_vals = f_exacta(x, epsilon)
f_approx_vals = f_approx(x, epsilon, h)

# Graficar ambas derivadas
plt.figure(figsize=(10, 6))
plt.plot(x, f_exacta_vals, label="Derivada exacta", color="blue")
plt.plot(x, f_approx_vals, label="Derivada aproximada", linestyle="--", color="red")
plt.title("Comparación entre la derivada exacta y la aproximada")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.legend()
plt.grid(True)
plt.show()

# Parte (b): Probar el programa con n=10 y ε=1/5 (esto ya está cubierto en el código)
```

c) Ajustar el valor de n hasta que la diferencia máxima entre la derivada exacta y la aproximada sea menor que 0.1

```python
# Parte (c): Encontrar el valor de n tal que la diferencia máxima sea menor que 0.1
diferencia_maxima = 0.1
while True:
    x = np.linspace(0, 1, n + 1)[1:]  # Recalculamos x excluyendo el 0
    h = x[1] - x[0]  # Recalcular h

    f_exacta_vals = f_exacta(x, epsilon)
    f_approx_vals = f_approx(x, epsilon, h)

    # Calcular la diferencia máxima
    max_diff = np.max(np.abs(f_exacta_vals - f_approx_vals))

    if max_diff < diferencia_maxima:
        break

    n += 1

print(f"El valor adecuado de n es: {n}")
```

```python
import numpy as np
from scipy.special import roots_legendre

# Función que queremos integrar
def f(x):
    return np.exp(4 * x)

# Método de Gauss-Legendre para calcular la integral
def gauss_legendre_integration(f, a, b, n):
    # Obtener las raíces (puntos) y los pesos de la cuadratura de Gauss-Legendre
    puntos, pesos = roots_legendre(n)

    # Transformar los puntos al intervalo [a, b]
    puntos_transformados = 0.5 * (b - a) * puntos + 0.5 * (b + a)

    # Aplicar la fórmula de cuadratura
    integral = 0.5 * (b - a) * np.sum(pesos * f(puntos_transformados))

    return integral

# Valor exacto de la integral
a_exacto = (np.exp(4) - 1) / 4

# Función para calcular la integral y comparar con el valor exacto
def calcular_integral_con_precision(f, a, b, epsilon):
    n = 1  # Empezamos con n=1
    while True:
        integral_aproximada = gauss_legendre_integration(f, a, b, n)
        error = abs(a_exacto - integral_aproximada)

        if error <= epsilon:
            return n, integral_aproximada, error
        n += 1

# Parámetros
a = 0
b = 1

# Parte (a): ε = 1/100
epsilon_1 = 1 / 100
n_1, integral_1, error_1 = calcular_integral_con_precision(f, a, b, epsilon_1)
print(f"Para ε = {epsilon_1}, n = {n_1}, integral = {integral_1}, error = {error_1}")

# Parte (b): ε = 1/1000
epsilon_2 = 1 / 1000
n_2, integral_2, error_2 = calcular_integral_con_precision(f, a, b, epsilon_2)
print(f"Para ε = {epsilon_2}, n = {n_2}, integral = {integral_2}, error = {error_2}")

# Parte (c): ε = 1/10000
epsilon_3 = 1 / 10000
n_3, integral_3, error_3 = calcular_integral_con_precision(f, a, b, epsilon_3)
print(f"Para ε = {epsilon_3}, n = {n_3}, integral = {integral_3}, error = {error_3}")
print(f"Para que el error sea menor o igual que {epsilon}, n debe ser de al menos {n}")
```

```python
from scipy.special import roots_legendre

# Función a integrar: e^(-x^2)
def f(x):
    return np.exp(-x**2)

# Método de Gauss-Legendre para calcular la integral
def gauss_legendre_integration(f, a, b, n):
    # Obtener las raíces y los pesos de Gauss-Legendre
    puntos, pesos = roots_legendre(n)

    # Transformar los puntos al intervalo [a, b]
    puntos_transformados = 0.5 * (b - a) * puntos + 0.5 * (b + a)

    # Aplicar la cuadratura
    integral = 0.5 * (b - a) * np.sum(pesos * f(puntos_transformados))

    return integral

# Función para calcular la aproximación de T(n, L)
def T(n, L):
    # Aproximamos la integral 2 * ∫_0^L e^(-x^2) dx
    return 2 * gauss_legendre_integration(f, 0, L, n)

# Valores de L que queremos usar
L_values = [2, 4, 6, 8, 10]
# Valores de n que queremos usar
n_values = [100, 200, 300, 400, 500]

# Valor exacto de la integral (sqrt(pi))
exact_value = np.sqrt(np.pi)

# Generar tabla para los valores de T(n, L)
print("Tabla de T(n, L):")
for n in n_values:
    row = [f"n = {n}"]
    for L in L_values:
        T_value = T(n, L)
        row.append(f"T({n}, {L}) = {T_value:.6f}")
    print("\t".join(row))

# Generar tabla de errores
print("\nTabla de errores en T(n, L):")
for n in n_values:
    row = [f"n = {n}"]
    for L in L_values:
        T_value = T(n, L)
        error = abs(T_value - exact_value)
        row.append(f"Error({n}, {L}) = {error:.6e}")
    print("\t".join(row))
```

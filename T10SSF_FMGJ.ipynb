Use la integración MOnte Carlo para hallar el volumen de un hemisferio de radio 1.

Podemos usar la simetría del problema para encontrar el volumen de un cuarto (el cuarto positivo (x, y, z)) del hemisferio, luego multiplicarlo por 4 para obtener nuestra respuesta final. Para el volumen conocido que rodea nuestra pieza desconocida, usemos el cubo unitario: z, y y z varían de 0 a 1. Este rango de valores corresponde exactamente a los números aleatorios dadios por random.random().

```python
from random import random
from math import sqrt
```

```python
#Como el radio es 1, podemos dejarlo fuera de la mayor parte de cálculos.
N= 1000000 #Numero de puntos aleatorios en la unidad del cubo.
count = 0 #Numero de puntos en la esfera.
for j in range (N):
  point = (random(), random(), random())
  if point[0] * point[0] + point[1] + point[2]* point[2] <= 1:
    count= count+1
#print(4 * count / N)

R= float(count)/float(N)
R= R * 4 #El volumen real está en nuestra prueba.
print('El volumen del hemisferio de radio 1 es %0.4f +/- %0.4f',
      (R, 4*sqrt(N)/float(N)))
```

La integración por Monte Carlo es una técnica que usa muestras aleatorias para estimar el valor de una integral. Para este caso fue el volumen del hemisferio de radio 1.

Lo que se hace en este código es generar puntos aleatorios y checar si se encuentran dentro del hemisferio. Luego, se calcula la relación de puntos dentro del hemisferio, lo que representa la fracción del volumen del cubo ocupado por el hemisferio.

Calcular $∫_{0}^{π}sen(x)dx$ usando el método Monte Carlo. Reporta la incertidumbre en el resultado y compárala con el resultado conocido.

```python
import numpy as np
```

```python
#Num de muestras
N=10000
#Generaremos los puntos aleatorios en el intervalo
x_random= np.random.uniform(0, np.pi, N)
#Ahora evaluamos la función a integrar en los puntos generados
f_v= np.sin(x_random)
#Estimamos la integral como el promedio de los valores arrojados
int_e=(np.pi-0)*(np.mean(f_v))
#Calculamos la incertidumbre:
incertidumbre= (np.pi-0)* np.std(f_v)/np.sqrt(N)
#Valor analítico de la integral para comparar. Sabemos que es 2
valor_int= -np.cos(np.pi)+np.cos(0)
print(f'Estimación de la integral: {int_e:.5f}')
print(f'Incertidumbre: {incertidumbre:.5f}')
print(f'Valor analítico de la integral: {valor_int}')
print(f"Diferencia respecto al valor verdadero: {abs(int_e - valor_int):.5f}")
```

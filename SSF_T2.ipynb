```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
```

```python
#Comencemos con el ejercicio 1, inciso a) definiendo la funcion a integrar
def fx(x):
    funcion = (1/(mt.sqrt(2*mt.pi)))*(mt.exp((-x**2)/2))
    return funcion
```

Método del Trapecio

```python
#Ahora definamos la funcion para poder aplicar el metodo del trapecio para 6 y 15 intervalos/
def tr(a,b,n, func, valorteorico): #Donde a y b son los limites superior e inferior.
#n es el numero de intervalos y la funcion a integrar
    suma = []
    intv = np.linspace (a, b, n+1)
    suma.append(float(func(intv[0])))
    suma.append(float(func(intv[-1])))
    h= (b-a)/n
    for i in (intv[1:-1]):
      suma.append(2*func(float(i)))
    result = (h/2)*(np.sum(suma))
    error = abs(result-valorteorico)/(valorteorico)
    return float(result), float(error)
```

```python
print("Para el método del trapecio:")
print({tr(-1,1,6,fx,0.6826894921)}, "Para 6 intervalos y error relativo respectivamente.")
print({tr(-2,0,15,fx, 0.06826894921)}, "Para 15 intervalos y error relativo.")
```

Método Simpson

```python
#Para el metodo Simpson escribimos la difinicion matematicamente.
def simpson(a, b, n, func, valorteorico):
  h= (b-a)/n
  intv = np.linspace(a,b,n+1)
  func = np.vectorize(func)
  S = func(intv[0]) + func(intv[-1]) + 4*np.sum(func(intv[1:-1:2])) + 2*np.sum(func(intv[2:-2:2]))
  #Lo que estamos haciendo con lo anterior es definir los intervalos pares y los impares
  results = (h/3)*S
  error = abs((results - valorteorico)/valorteorico)
  return float(results), float(error)
```

```python
print("Para el método Simpson")
print({simpson(-1,1,6,fx,0.6826894921)}, "Para 6 intervalos y error relativo respectivamente.")
print({simpson(-1,1,15,fx,0.6826894921)}, "Para 15 intervalos y error relativo.")
```

Método de Gauss-Lengendre

```python
def g_l(a,b,n,func,valorteorico):
  xi, wi = np.polynomial.legendre.leggauss(n) #Esto para obtener los puntos y los pesos para n.
  ent= (xi/2)*(b-a)+((b+a)/2)
  func = np.vectorize(func)
  resultgl = ((b-a)/2)*np.sum(wi*func(ent))
  error= abs((resultgl-valorteorico)/valorteorico)
  return float(resultgl), float(error)
```

```python
print("Por el método de Gauss-Legendre:")
print({g_l(-1,1,6,fx, 0.6826894921)}, "Para 6 intervalos y error relativo respectivamente.")
print({g_l(-1,1,15,fx, 0.6826894921)}, "Para 15 intervalos y error relativo")
```
